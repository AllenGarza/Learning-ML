Working with large amounts of data

```python

url = 'https://data.cdc.gov/resource/ks3g-spdg.csv'

df = pd.read_csv(url, error_bad_lines=False)

## we can define a newdf with the same columns 
## In doing this, we cun chunk and filter out our data every iteration

new_df = pd.DataFrame(columns=df.columns)

for df in pd.read_csv(url, error_bad_lines=False, chunksize=5):
    results = df.groupby(['race_and_hispanic_origin']).count()
    
    new_df = pd.concat([new_df, results])
    print("a chunk\n")
    print(new_df)
```
