# Practicing plotting and graphing data with Seaborn

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib as plt

%matplotlib inline
%reload_ext autoreload
%autoreload 2
```

# Import Data

```python
df = sns.load_dataset('car_crashes')
df
```

Distribution plots

```python
#use shift+tab with cursor on code in jupyter to get info
## Distribution plot
sns.displot(df['not_distracted'], kde=False, bins=25)

## Joint plots
sns.jointplot(x='speeding', y='alcohol', data=df, kind='kde')
# kind='kde' kernal data estimation (kde) estimates distributions of data
# kind='hex' distribution using hex shapes
```

```python
##KDE Plot
sns.kdeplot(df['alcohol'])
```

```python
##pairplots
sns.pairplot(df)
```

```python
df = sns.load_dataset('tips')

sns.pairplot(df, hue='sex', palette='Blues')
```

```python
## Rug Plot
# pulls single data points of a specific column as sticks, density implies common
sns.rugplot(df['tip'])
```

```python
## Styling
sns.set_style('white')

sns.jointplot(x='speeding', y='alcohol', data=df)
```
